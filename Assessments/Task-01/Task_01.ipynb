Write a program to check whether a given year is a leap year

```python
import calendar
year = int(input("Enter a year: "))
if len(str(year))==4:
    if calendar.isleap(year):
        print(f"{year} is a Leap Year")
    else:
        print(f"{year} is Not a Leap Year")
else:
    print(f"{year} is invalid! enter valid year")
```

Write a program to check whether a number is divisible by both 5 and 11

```python
number=int(input("enter the value:"))
if number%5==0 and number%11==0:
    print(f"{number} the number is divisible by both 5 and 11")
else:
    print(f"{number} the number is not divisible by  both 5 and 11")
```

Write a program to calculate the sum of first N natural 

```python
n=int(input("enter:"))
sum=0
if n>1:
    for i in range(1,n+1):
        sum+=i
    print(sum)
else:
    print("invalid")
```

Write a program to print the multiplication table of a using while loop

```python
n=int(input("enter the table:"))
upto=int(input("enter the value upto the table want to print:"))
total = 0
count = 1
while count <= upto:
   total = count*n
   print(f"{count}x{n}={total}")
   count += 1
    
    
```

Print a pyramid pattern as *

```python
size = int(input("enter the row:"))
for i in range(size):
   for j in range(1, size - i):
       print(" ", end="")
   for k in range(0, i + 1):
       print("*", end=" ")
   print()
    
```
